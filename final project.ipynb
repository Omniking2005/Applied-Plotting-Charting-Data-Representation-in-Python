```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

from datetime import datetime 

qq = pd.read_csv('Book1.csv')
ww = pd.read_csv('book 2.csv')
qq = qq.dropna()

wq = ww[['NO2']]

wq['NO2'] = wq['NO2'].astype(float)
qw = qq[['NO2','nd']]
papa1 = qw.iloc[:,1]
mama1 = qw.iloc[:,0]



plt.plot(wq,label = 'REGULAR DAY')
plt.scatter(papa1,mama1, c = 'red', label = 'RELIGIOUS DAY')
plt.ylim(0, 275)
plt.legend(loc = 0, frameon=True, title='DAY TYPE', fontsize=6)
plt.xticks( np.linspace(15,15 + 30*8 , num = 8), (r'January', r'Febuary', r'March', r'April', r'May', r'June', r'July', r'August') )
plt.yticks( np.linspace(12.5 , 12.5 + 25*11 , num =11 ), (25,50,75,100,125,150,175,200,225,250,275))
plt.ylabel('AMOUNT OF NO2 IN AIR')
plt.title(r'AMOUNT OF NO2 IN AIR "ON REGULAR DAY" vs "ON RELIGIOUS EVENTS" IN Ahmedabad  ')


plt.show()





```
